```python
import sys
import pandas as pd
sys.path.append('D:/research/jokesclustering')# ваш путь до корня проекта
sys.path.append('D:/research/jokesclustering/vector_clustering/models/comps/')
from components_simple_norm import SimpleNormVectorizer
from model import HardClustering
from vector_clustering.data.manager import get_jokes_as_dataframe, ARANEUM_FASTTEXT
```

### Инициализация

```python
jokes_frame = get_jokes_as_dataframe()
texts = jokes_frame.joke_text.values
```

```python
vectorizer = SimpleNormVectorizer(ARANEUM_FASTTEXT)
vectors = vectorizer.fit(texts)
```

```python
model = HardClustering(texts=texts, 
                       vectors=vectors, 
                       initial_threshold=0.9, 
                       number_target_clusters=50, 
                       random_state=42)
```

### Запуск

```python
model.custom_cycles(end_sim=0.7)
```

```python
model.display_current_cycle()
```

```python
c, _ = model.get_current_clusters()
```

```python
for i, c_ in enumerate(c):
    print('Кластер {}, размер: {}'.format(i, c_[1]))
```
